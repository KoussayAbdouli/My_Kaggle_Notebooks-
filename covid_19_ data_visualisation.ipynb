```python
def hospitalisation(df):
    if df['Patient addmited to regular ward (1=yes, 0=no)'] == 1:
        return "surveillance"
    elif df['Patient addmited to semi-intensive unit (1=yes, 0=no)'] == 1:
        return 'soins semi-intensives'
    elif df['Patient addmited to intensive care unit (1=yes, 0=no)'] == 1:
        return 'soins intensifs'
    else:
        return 'inconnu'
   
```

```python
df['statut'] = df.apply(hospitalisation , axis= "columns")
```

```python
plt.figure(figsize=(12,8))
df['statut'].value_counts().plot.pie()
```

```python
df[blood_columns]
```

```python
for col in blood_columns:
    plt.figure()
    for i in df['statut'].unique():
        sns.distplot(df[df['statut'] == i][col] , label = i )
    plt.legend()
```

### Hypothés : T-Test

```python
from scipy.stats import ttest_ind
```

```python
df_positive.shape
```

```python
df_negative.shape
```

```python
balanced_neg = df_negative.sample(df_positive.shape[0])
```

```python
def t_test(col):
    alpha = 0.5
    stat , p = ttest_ind(balanced_neg[col].dropna(), df_positive[col].dropna())
  
    if p < alpha:
        return 'H0 Rejetée'
    else:
        return 0 
```

```python
for col in blood_columns:
    print(f'{col:-<50}{t_test(col)}')
```
